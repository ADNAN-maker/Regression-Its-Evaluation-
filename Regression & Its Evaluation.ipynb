Q1. What is Simple Linear Regression?
***
Ans:- Simple linear regression is a statistical method used to model the relationship between one independent variable and one dependent variable using a straight line. It aims to find the best-fitting line that describes how the dependent variable changes as the independent variable changes.

Key Concepts:
1.   Independent Variable
2.   Dependent Variable
3.   Linear Relationship
4.   Regression Line

Purpose:

*   Prediction: To predict the value of the dependent variable for a given value of the independent variable.
*   Understanding Relationships: To understand how changes in the independent variable affect the dependent variable.
*   Identifying Trends: To identify trends and patterns in data.

Q2. What are the key assumptions of Simple Linear Regression?
***
Ans:- The key assumptions of simple linear regression are linearity, independence, homoscedasticity, and normality. These assumptions ensure that the model accurately reflects the relationship between the independent and dependent variables and that the results are reliable.
1.   Linearity: A linear relationship exists between the independent variable (X) and the dependent variable (Y).
2.   Independence: The residuals (the differences between observed and predicted values) are independent of each other.
3.   Homoscedasticity: The variance of the residuals is constant across all levels of the independent variable.
4.   Normality: For any given value of X, the dependent variable Y is normally distributed. This means that the distribution of the errors is also normal.

Q3. What is heteroscedasticity, and why is it important to address in regression
models?
***
Ans:- Heteroscedasticity in regression models refers to the situation where the variance of the residuals (or error terms) is not constant across all levels of the independent variables, violating a key assumption of ordinary least squares (OLS) regression, known as homoscedasticity. Addressing heteroscedasticity is crucial because it can lead to inefficient and unreliable regression coefficient estimates, making the model's inferences inaccurate and potentially misleading.

importantance
1.  Violation of OLS Assumptions:
Standard linear regression assumes that the variance of the errors is constant (homoscedasticity). When heteroscedasticity is present, this assumption is violated, meaning the OLS estimates are no longer BLUE (Best Linear Unbiased Estimators).
2.  Inefficient Estimates:
While heteroscedasticity does not bias the coefficient estimates, it makes them inefficient. The standard errors of the coefficients will be underestimated or overestimated, leading to incorrect p-values and confidence intervals.
3. Unreliable Hypothesis Testing:
If the standard errors are incorrect, then hypothesis tests (e.g., t-tests for individual coefficients) will not be accurate, potentially leading to incorrect conclusions about the statistical significance of independent variables.
4. Model Interpretation Issues:
The uneven spread of residuals in a heteroscedastic model can indicate that the model is not adequately capturing the relationships in the data, especially when the variance of the errors changes systematically with the independent variables.

Q4. What is Multiple Linear Regression?
***
Ans:- Multiple linear regression is a statistical method that analyzes the relationship between one dependent variable and two or more independent variables. It aims to find the best-fitting linear equation that predicts the dependent variable based on the values of the independent variables. Essentially, it's an extension of simple linear regression, allowing for more than one predictor in the model.

Dependent Variable:
The variable you are trying to predict (e.g., house price).
Independent Variables:
The variables used to predict the dependent variable (e.g., square footage, number of bedrooms, location).

Formula:
The general form of a multiple linear regression equation is:

Y = b₀ + b₁X₁ + b₂X₂ + ... + bₙXₙ + ε

Where:
*  Y is the dependent variable.
*  X₁, X₂, ... Xₙ are the independent variables.
*  b₀ is the intercept (the value of Y when all independent variables are zero).
*  b₁, b₂, ... bₙ are the coefficients (slopes) for each independent variable.
*  ε is the error term (the difference between the predicted and actual values).

Q5. What is polynomial regression, and how does it differ from linear
regression?
***
Ans:- Polynomial regression is a type of regression analysis that models the relationship between a dependent variable and an independent variable as an nth-degree polynomial. It differs from linear regression primarily in its ability to model non-linear relationships between variables, where linear regression is restricted to fitting a straight line to the data.

Key Differences between Polynomial and Linear Regression:

Relationship Modeling:

Linear Regression: Assumes a linear relationship between the independent and dependent variables, fitting a straight line to the data. Its equation is typically of the form y = mx + c.

Polynomial Regression: Models a curved relationship by incorporating higher-order terms of the independent variable (e.g., x², x³, etc.). This allows it to fit data with non-linear patterns, like curves or bends.

Q6. Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
* X = [1, 2, 3, 4, 5]
* Y = [2.1, 4.3, 6.1, 7.9, 10.2]

Plot the regression line over the data points.
***

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

model = LinearRegression()

model.fit(X, Y)

Y_pred = model.predict(X)

plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()

print(f"Intercept: {model.intercept_}")
print(f"Coefficient: {model.coef_[0]}")
```

Q7. Fit a Multiple Linear Regression model on this sample data:
* Area = [1200, 1500, 1800, 2000]
* Rooms = [2, 3, 3, 4]
* Price = [250000, 300000, 320000, 370000]

Check for multicollinearity using VIF and report the results.
***


```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}

df = pd.DataFrame(data)

X = df[['Area', 'Rooms']]
y = df['Price']

model = LinearRegression()
model.fit(X, y)

print("Multiple Linear Regression Model:")
print(f"Intercept: {model.intercept_}")
print("Coefficients:")
for col, coef in zip(X.columns, model.coef_):
    print(f"  {col}: {coef}")

print("\nChecking for Multicollinearity (VIF):")
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)

print("\nVIF Interpretation:")
print("- VIF = 1: No correlation between the independent variable and other variables.")
print("- 1 < VIF < 5: Moderately correlated.")
print("- VIF >= 5: Highly correlated (potentially problematic).")
```

Q8. Implement polynomial regression on the following data:
* X = [1, 2, 3, 4, 5]
* Y = [2.2, 4.8, 7.5, 11.2, 14.7]

Fit a 2nd-degree polynomial and plot the resulting curve.
***


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)

model = LinearRegression()

model.fit(X_poly, Y)

Y_pred = model.predict(X_poly)

plt.scatter(X, Y, color='blue', label='Data Points')

X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_range_poly = poly_features.transform(X_range)
Y_range_pred = model.predict(X_range_poly)
plt.plot(X_range, Y_range_pred, color='red', linewidth=2, label='Polynomial Regression Curve (Degree 2)')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression (Degree 2)')
plt.legend()
plt.grid(True)
plt.show()

print("Polynomial Regression Model (Degree 2):")
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")
```

Q9. Create a residuals plot for a regression model trained on this data:
* X = [10, 20, 30, 40, 50]
* Y = [15, 35, 40, 50, 65]

Assess heteroscedasticity by examining the spread of residuals.
***

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

model = LinearRegression()
model.fit(X, Y)

Y_pred = model.predict(X)

residuals = Y - Y_pred

plt.scatter(Y_pred, residuals, color='blue')
plt.hlines(y=0, xmin=Y_pred.min(), xmax=Y_pred.max(), color='red', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.grid(True)
plt.show()

print("Assessing Heteroscedasticity:")
print("Examine the residuals plot. If the points are randomly scattered around the horizontal line at 0 with no discernible pattern (like a cone shape), it suggests homoscedasticity. If the spread of residuals changes systematically with the predicted values, it suggests heteroscedasticity.")
```

Q10. Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.
***
Ans:- When dealing with heteroscedasticity and multicollinearity in a real estate price prediction model, a data scientist would take a series of steps to diagnose, address, and validate their approach. Here's a breakdown of those steps:

Addressing Heteroscedasticity:

Heteroscedasticity means the variance of the errors is not constant across all levels of the independent variables. This doesn't bias the coefficient estimates, but it makes them inefficient and invalidates standard errors, p-values, and confidence intervals.

Diagnosis:
Residuals Plot: As we saw earlier, plotting residuals against predicted values (or independent variables) is a primary way to visually detect heteroscedasticity. A fanning-out or cone shape indicates increasing variance.
Statistical Tests: Formal tests like the Breusch-Pagan test or the White test can statistically assess the presence of heteroscedasticity.
Addressing Heteroscedasticity:
Transformations: Applying transformations to the dependent variable (e.g., taking the logarithm of house prices) can often stabilize the variance. Other transformations like square root or inverse might also be considered depending on the nature of the heteroscedasticity.
Weighted Least Squares (WLS): If the pattern of heteroscedasticity is known or can be estimated, WLS can be used. This method assigns lower weights to observations with higher variance and higher weights to observations with lower variance, effectively giving more importance to the more reliable data points.
Robust Standard Errors: Using robust standard errors (also known as Heteroscedasticity-Consistent Standard Errors or HCSE) is a common approach. These standard errors are calculated in a way that is less sensitive to heteroscedasticity, providing more reliable p-values and confidence intervals even if the heteroscedasticity is not explicitly modeled. This doesn't change the coefficient estimates themselves but corrects their standard errors.
Revisit Model Specification: Sometimes, heteroscedasticity can be a sign that the model is missing important variables or that the functional form of the relationship is incorrect. Adding relevant variables or considering non-linear relationships might help.
Addressing Multicollinearity:

Multicollinearity occurs when independent variables in a regression model are highly correlated with each other. This makes it difficult to determine the individual effect of each correlated variable on the dependent variable, leading to unstable and less reliable coefficient estimates.

Diagnosis:
Correlation Matrix: Calculating a correlation matrix for the independent variables can reveal strong pairwise correlations.
Variance Inflation Factor (VIF): As we saw in the previous example, VIF quantifies how much the variance of the estimated regression coefficient is increased due to multicollinearity. A high VIF (often > 5 or > 10, depending on the context) for a variable indicates significant multicollinearity.
Addressing Multicollinearity:
Remove one of the correlated variables: If two or more variables are highly correlated, you can remove one of them from the model. The choice of which variable to remove might depend on domain knowledge, statistical significance, or which variable is considered more theoretically important.
Combine correlated variables: Create a new variable that is a combination of the highly correlated variables. For example, if 'number of bedrooms' and 'number of bathrooms' are highly correlated, you could create a 'total number of rooms' variable (if appropriate for the context).
Principal Component Analysis (PCA): PCA is a dimensionality reduction technique that can transform the original correlated variables into a set of uncorrelated variables (principal components). These principal components can then be used as predictors in the regression model.
Ridge Regression or Lasso Regression: These are regularization techniques that can help to reduce the impact of multicollinearity by adding a penalty term to the regression objective function. This can shrink the coefficients of correlated variables towards zero, making the model more stable.
Increase Sample Size: In some cases, increasing the sample size can help to reduce the impact of multicollinearity, as it provides more information for the model to disentangle the effects of the correlated variables. However, this is not always a feasible solution.
Ensuring a Robust Model:

After addressing heteroscedasticity and multicollinearity, it's crucial to:

Re-evaluate Model Assumptions: Check the assumptions of the regression model again with the adjusted model (e.g., using residuals plots, normality tests).
Validate the Model: Use techniques like cross-validation to assess how well the model generalizes to new, unseen data.
Interpret Results Carefully: Understand the limitations of the model and interpret the coefficients in the context of the chosen approach (e.g., understanding that transformations change the interpretation).
Consult Domain Expertise: Work with real estate experts to ensure that the model makes sense from a business perspective and that the chosen variables and their relationships are meaningful.
